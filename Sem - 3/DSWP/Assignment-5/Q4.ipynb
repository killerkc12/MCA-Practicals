4. We have a dataset “Mall_Customers.csv” which contains data of customers
who frequently visits a mall. The features of dataset like Customer ID, Gender,
Age, Annual Income, and Spending score (1-100) assigned by Mall based on
average spending by the customer. Create Hierarchical Clustering Model to
segment the customers in different clusters based on the features so that Mall
Owner can target the customer from similar group.
a. Analyze the different features of data set.
b. Find the optimum value of k (using dendrogram)
c. Create Hierarchical Clustering Model.
d. Visualize the number of clusters.
e. Analyze the different segments of customers.

```python
# a. Analyze the different features of data set.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,[3,4]]
```

```python
X.head()
```

```python
# b. Find the optimum value of k (using dendrogram)
import scipy.cluster.hierarchy as sch
dendrogrm = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
plt.title("Dendrogram")
dendrogrm = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.axhline(y=250,color='r',LineStyle="--")
```

```python
# c. Create Hierarchical Clustering Model.
from sklearn.cluster import AgglomerativeClustering
```

```python
hc= AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_pred= hc.fit_predict(X)
y_pred
```

```python
# d. Visualize the number of clusters.
plt.scatter(X.iloc[:,0],X.iloc[:,1],s=100, marker='.',c=y_pred)
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
